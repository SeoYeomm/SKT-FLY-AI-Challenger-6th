## Lineplot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
sns.lineplot(x=x, y=y)
sns.lineplot(x=x, y=z)
plt.title('sin(x) and cos(x)')
plt.show()
```

## Bar Chart

```python
cat = np.array(['A', 'B', 'C', 'D'])
value = np.array([3, 8, 1, 10])

sns.barplot(x=cat, y=value, hue=cat)
for i,v in enumerate(value):
  plt.text(i, v, str(v), fontsize = 15, ha = 'center')

plt.show()
```

```python
cat = np.array(['A', 'B', 'C', 'D'])
value = np.array([3, 8, 1, 10])

sns.barplot(x=cat, y=value, palette = 'viridis')
for i,v in enumerate(value):
  plt.text(i, v, str(v), fontsize = 15, ha = 'center')

plt.show()
```

```python
cat = np.array(['A', 'B', 'C', 'D'])
value = np.array([3, 8, 1, 10])
value.sort()

sns.barplot(x=value,y=cat)
plt.show()
```

## Count Plot

```python
data = ['A','B','A','C','B','C','A']
sns.countplot(x=data)
plt.show()
```

## Histogram


```python
data = np.random.randn(1000)
sns.histplot(data, bins=40, kde = True) # 막대개수 40개
plt.show()
```

## Scatterplot

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.scatterplot(x=df['sepal_length'], y=df['sepal_width'])
```

```python
df.columns
```

```python
sns.scatterplot(x='petal_length', y='petal_width',data=df, hue='species', style='species')
# 3개의 축 (x, y, 색상) -> 2차원이지만
```

```python
sns.pairplot(df, hue='species') # 모든 경우의 수 다 그려줌
```

## Box plot

```python
sns.boxplot(data=df, x='species', y='petal_length', hue='species')
plt.show()
```

## Violinplot

```python
sns.violinplot(data=df, x='species', y='petal_length', hue='species')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
sns.violinplot(data=df, x='species', y='petal_length', hue='species')
plt.show()
```
