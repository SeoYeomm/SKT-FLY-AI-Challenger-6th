# 데이터 집계 및 그룹화

```python
data = {
    'Name':['Alice','Bob','Charlie','Jane'],
    'Age':[25,30,28,40],
    'Gender':[1, 0, 1, 0]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby('Gender')['Age'].sum() # Gender별로 묶어서 age의 합
```

```python
df.groupby('Gender')['Age'].mean() # Gender별로 묶어서 age의 평균
```

```python
data = {
    'Name':['Alice','Bob','Charlie','Jane'],
    'Age':[25,30,28,40],
    'Gender':[1, 0, 1, 0],
    'Area': ['A','B','A','B']
}
df = pd.DataFrame(data)
df
```

```python
df.groupby(['Gender', 'Area'])['Age'].sum() # Gender and Area별로 묶어서 age의 평균
```
