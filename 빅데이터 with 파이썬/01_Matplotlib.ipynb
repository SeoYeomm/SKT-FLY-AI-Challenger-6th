# Matplotlib

## 선형 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.plot(x, y, label = 'Score')
plt.title("Simple Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend() # 범례
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4])
y =  np.array([0, 2, 4, 6, 8])
z = x**2

plt.plot(x, y, label = 'Score')
plt.plot(x, z, label = 'Square')
plt.title("Simple Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend() # 범례
plt.show()
```

## Bar Chart

```python
x = ['A','B','C','D']
y = [10, 30, 20, 40]

plt.bar(x, y)
plt.show()
```

```python
plt.style.use('ggplot')

x = ['A','B','C','D']
y = [10, 30, 20, 40]

plt.bar(x, y)
for i,v in enumerate(y):
    plt.text(i, v, str(v),ha='center')
plt.ylim([0,50])
plt.show()
```

## 히스토그램 (Histogram)

```python
x = np.random.randn(1000)

plt.hist(x)
plt.show()
```

## 산점도 (Scatter Plot)

```python
x = [5, 7, 8, 9, 11, 4, 9]
y = [99, 86, 75, 66, 65, 99, 75]

plt.scatter(x,y)
plt.show()
```

## 박스 플롯 (Box Plot)

```python
# boxplot: 데이터 범위 + 중앙값 (이상치 판단에 좋음)
x = [99, 86, 75, 66, 65, 99, 75]
plt. boxplot(x)
plt.show()
```

## 다중 플롯과 서브 플롯

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
z = [0, -1, -4, -9, -16]

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,z)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,z)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(x,y)
axes[0,1].plot(x,z)
axes[1,0].plot(x,y)
axes[1,1].plot(x,z)
plt.show()
```
